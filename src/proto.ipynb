```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.smbc-comics.com/'
```

```python
raw = requests.get(url)
if raw.status_code == 200:
    print('Request successful')
else:
    print('Request failed with status code:', raw.status_code)
```

```python
soup = bs(raw.text, 'html.parser')
```

```python
# get the main comic image
comic_div = soup.find('div', id='cc-comicbody')
comic_img = comic_div.find('img')
comic_url = comic_img['src']
comic_title = comic_img['title']
print('Comic URL:', comic_url)
print('Comic Title:', comic_title)
```
